```python
import pandas as pd
import sqlalchemy
from binance import Client, BinanceSocketManager
```

```python
api_key = ''
api_secret = ''
```

```python
client = Client(api_key, api_secret)
```

```python
bsm = BinanceSocketManager(client)
```

```python
socket = bsm.trade_socket('ETHUSDT')
```

```python
engine = sqlalchemy.create_engine('sqlite:///ETHUSDTstream.db')
```

```python
def create_dataframe(msg):
    df = pd.DataFrame([msg])
    if 'Price' in df.columns:
        df = df.loc[:, ['Symbol', 'Time', 'Price']]
    elif 'Price' not in df.columns:
        df = df.loc[:, ['s', 'E', 'p']]
        df.columns = ['Symbol', 'Time', 'Price']
    df.Price = df.Price.astype(float)
    df.Time = pd.to_datetime(df.Time, unit='ms')
    return df
```

```python
while True:
    await socket.__aenter__()
    msg = await socket.recv()
    await socket.__aexit__(None, None, None)
    try:
        frame = create_dataframe(msg)
        frame.to_sql('ETHUSDT', engine, if_exists='append', index=False)
        print(frame)
    except:
        print(f'Error: {msg["m"]}')
```
